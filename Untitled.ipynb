```python
import pandas as pd
```

```python
database = pd.read_excel("gruppi_DNA.xlsx")
```

```python
database.iloc[:,0:28].dropna().tail()
```

```python
df_new = database.iloc[:,0:28].dropna()
```

```python
df_new.describe()
```

```python
df_new.iloc[:,1:] = df_new.iloc[:,1:]/100
```

```python
import matplotlib.pyplot as plt
```

```python
df_new_unlab = df_new.iloc[:,1:]
```

```python
plt.figure(figsize=(10, 8))
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 30):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df_new_unlab)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 30), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 30, init = 'k-means++', random_state = 42)
kmeans.fit(df_new_unlab)
```

```python
y_kmeans = kmeans.fit_predict(df_new_unlab)
```

```python
y_kmeans1=y_kmeans
y_kmeans1=y_kmeans+1
cluster = pd.DataFrame(y_kmeans1)
df_new_2 = df_new_unlab*100
df_new_2['cluster'] = cluster
df_new_2["nation"] = df_new["Unnamed: 0"]
kmeans_mean_cluster = pd.DataFrame(round(df_new_2.groupby('cluster').mean()*100,1))
kmeans_mean_cluster
```
